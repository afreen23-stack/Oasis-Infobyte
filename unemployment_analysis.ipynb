```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Unemployment in India.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
data.columns = data.columns.str.strip()
data.columns
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=data)
plt.xticks(rotation=90)
plt.title("Unemployment Rate by State")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data)
plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
corr = data[['Estimated Unemployment Rate (%)',
             'Estimated Employed',
             'Estimated Labour Participation Rate (%)']].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Area', y='Estimated Unemployment Rate (%)', data=data)
plt.title("Unemployment Rate by Area")
plt.show()
```
